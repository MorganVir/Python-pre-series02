# Python - Pre-series02 - Sinusoidal Signals

### What is a sinusoid?  
It's a smooth periodic oscillation described by a sine or cosine function.  

### What physical phenomenon produces a sinusoid?  
It's produced by an uniform circular motion projected on one axis.  

### Diagram of a sine curve  
- x-axis: time (s)  
- y-axis: amplitude (UA here)  
- Remarkable points: maxima, minima, zero crossings, period T.  


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters 
A = 1.0            
f = 1.0            
omega = 2 * np.pi * f
phi = 0            
T = 1 / f          

#  Time vector 
t = np.linspace(0, 2*T, 500)  

# Sine wave 
x = A * np.sin(omega*t + phi)


plt.figure(figsize=(8, 4))

# Make the sine wave line thicker so it stands out more and it's easier to read quickly 
plt.plot(t, x, linewidth=2, color="blue")

# Remarkable points 
# Instead of showing ALL zeros and extrema with text,
# we only highlight one "type" of each:

plt.scatter([T/4, 3*T/4], [A, -A], color="darkorange", zorder=5, label="max/min")
plt.scatter([0, T, 2*T], [0, 0, 0], color="green", zorder=5, label="zeros")

# Annotations 
# Instead of repeating "zero" multiple times,
# we annotate just ONE maximum and ONE minimum.
plt.annotate("max = A", xy=(T/4, A), xytext=(T/4, A+0.4),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("min = -A", xy=(3*T/4, -A), xytext=(3*T/4, -A-0.6),
             ha="center", arrowprops=dict(arrowstyle="->"))

# Period marker
# Keep the vertical line at T, but don’t overload with labels in the legend
plt.axvline(T, color="red", linestyle="--")
plt.text(T, -0.3, "T", ha="center", color="red")

# Styling improvements 
# Add a horizontal axis line to make the zeros more obvious
plt.axhline(0, color="black", linewidth=0.8)

# Simplify title (mention parameters directly in text instead of legend clutter)
plt.title("Sine Wave (Amplitude = 1, Period = T)")

# Axes labels
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Keep legend, but now only 2 items (clearer, not overwhelming)
plt.legend()

# Lighten the grid so it doesn’t compete visually
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```


### Equation of a sinusoid  

$x(t) = A \sin(\omega t + \phi)$  

Where:  
- $A$: amplitude (unit will vary)
- $\omega$: angular frequency (rad/s)  
- $t$: time (s)  
- $\phi$: phase shift (rad)  
- $x(t)$: instantaneous value (same as amplitude)  



### Definitions and units  

- **Frequency $f$**: number of oscillations per second, unit = hertz (Hz).  
  Relation: $f = \frac{1}{T}$  

- **Period $T$**: duration of one complete cycle, unit = second (s).  
  Relation: $T = \frac{1}{f}$  

- **Amplitude $A$**: maximum deviation from zero, unit = same as signal.  

- **Angular frequency (pulsation) $\omega$**: rate of oscillation, unit = rad/s.  
  Relation: $\omega = 2\pi f$  

- **Phase $\phi$**: initial offset of the oscillation, unit = rad.  




```python
# Parameters 
A = 1.0
f = 1.0
omega = 2 * np.pi * f
T = 1 / f

# Time vector
t = np.linspace(0, 2*T, 500)

# Sine and Cosine
x_sin = A * np.sin(omega*t)
x_cos = A * np.cos(omega*t)

plt.figure(figsize=(9, 5))

plt.plot(t, x_sin, linewidth=2, color="blue", label="Sine: A·sin(ωt)")
plt.plot(t, x_cos, linewidth=2, color="green", linestyle="--", label="Cosine: A·cos(ωt)")

# Minimal key markers (one example only!)
plt.scatter([T/4], [A], color="darkorange", zorder=5, label="sine max")
plt.scatter([0], [0], color="purple", zorder=5, label="sine zero")

# Period marker
plt.axvline(T, color="red", linestyle="--")
plt.text(T, -0.3, "T", ha="center", color="red")

# Simple phase note (short & close to cosine curve)
plt.text(0.2, 0.7, "Cosine leads sine\nby π/2", color="green")

# Styling 
plt.axhline(0, color="black", linewidth=0.8)  
plt.title("Sine vs Cosine (Phase Shift π/2)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```


### Difference between sine and cosine  

Sine:  
$x(t) = A \sin(\omega t)$  

Cosine:  
$x(t) = A \cos(\omega t)$  

The cosine is the sine shifted by $\pi/2$ in phase.  
Both have same amplitude and period but are shifted in time.  



### Why is a sinusoid periodic?  
It repeats identically after each period $T$.  



### Why is a sinusoid continuous?  
It is defined for all real time values without discontinuities.  



### Why is a sinusoid parametric, and with which parameters?  
It depends on amplitude ($A$), frequency ($f$), angular frequency ($\omega$), and phase ($\phi$).
