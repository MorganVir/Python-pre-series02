# Python - Pre-series02 - Sinusoidal Signals

### What is a sinusoid?  
It's a smooth periodic oscillation described by a sine or cosine function.  

### What physical phenomenon produces a sinusoid?  
It's produced by an uniform circular motion projected on one axis.  

### Diagram of a sine curve  
- x-axis: time (s)  
- y-axis: amplitude (UA here)  
- Remarkable points: maxima, minima, zero crossings, period T.  


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters 
A = 1.0            # Amplitude
f = 1.0            # Frequency (Hz)
omega = 2 * np.pi * f
phi = 0            # Phase (rad)
T = 1 / f          # Period (s)

#  Time vector 
t = np.linspace(0, 2*T, 500)  # two periods

# Sine wave 
x = A * np.sin(omega*t + phi)

#  Plot 
plt.figure(figsize=(8, 4))
plt.plot(t, x, label="x(t) = A sin(ωt + φ)", color="blue")

# Remarkable points
points_x = [0, T/4, T/2, 3*T/4, T]
points_y = [0, A, 0, -A, 0]

plt.scatter(points_x, points_y, color="darkorange", zorder=5)

# Annotations
plt.annotate("zero", xy=(0, 0), xytext=(0, 0.3),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("max = A", xy=(T/4, A), xytext=(T/4, A+0.4),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("zero", xy=(T/2, 0), xytext=(T/2, 0.3),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("min = -A", xy=(3*T/4, -A), xytext=(3*T/4, -A-0.6),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("zero", xy=(T, 0), xytext=(T, 0.3),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.axvline(T, color="red", linestyle="--", label="Period T")
plt.text(T, -0.3, "T", ha="center", color="red")

# Styling
plt.title("Sine Wave with Remarkable Points")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```


### Equation of a sinusoid  

$x(t) = A \sin(\omega t + \phi)$  

Where:  
- $A$: amplitude (unit will vary)
- $\omega$: angular frequency (rad/s)  
- $t$: time (s)  
- $\phi$: phase shift (rad)  
- $x(t)$: instantaneous value (same as amplitude)  



### Definitions and units  

- **Frequency $f$**: number of oscillations per second, unit = hertz (Hz).  
  Relation: $f = \frac{1}{T}$  

- **Period $T$**: duration of one complete cycle, unit = second (s).  
  Relation: $T = \frac{1}{f}$  

- **Amplitude $A$**: maximum deviation from zero, unit = same as signal.  

- **Angular frequency (pulsation) $\omega$**: rate of oscillation, unit = rad/s.  
  Relation: $\omega = 2\pi f$  

- **Phase $\phi$**: initial offset of the oscillation, unit = rad.  




```python
# Parameters 
A = 1.0
f = 1.0
omega = 2 * np.pi * f
T = 1 / f

# Time vector
t = np.linspace(0, 2*T, 500)

# Sine and Cosine
x_sin = A * np.sin(omega*t)
x_cos = A * np.cos(omega*t)

# Plot 
plt.figure(figsize=(9, 5))
plt.plot(t, x_sin, label="Sine: A·sin(ωt)", color="blue")
plt.plot(t, x_cos, label="Cosine: A·cos(ωt)", color="green", linestyle="--")

# Remarkable points (sine)
plt.scatter([0, T/4, T/2, 3*T/4, T],
            [0, A, 0, -A, 0],
            color="darkorange", zorder=5)

# Annotations for sine
plt.annotate("zero", xy=(0, 0), xytext=(0, 0.3),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("max (sine)", xy=(T/4, A), xytext=(T/4, A+0.4),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("zero", xy=(T/2, 0), xytext=(T/2, 0.3),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("min (sine)", xy=(3*T/4, -A), xytext=(3*T/4, -A-0.6),
             ha="center", arrowprops=dict(arrowstyle="->"))
plt.annotate("zero", xy=(T, 0), xytext=(T, 0.3),
             ha="center", arrowprops=dict(arrowstyle="->"))

# Period line
plt.axvline(T, color="red", linestyle="--", label="Period T")
plt.text(T, -0.3, "T", ha="center", color="red")

# Note about cosine
plt.text(0.15, 0.8, "Cosine leads sine by π/2", color="green")

# Styling
plt.title("Sine vs Cosine (Phase Shift π/2)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```


### Difference between sine and cosine  

Sine:  
$x(t) = A \sin(\omega t)$  

Cosine:  
$x(t) = A \cos(\omega t)$  

The cosine is the sine shifted by $\pi/2$ in phase.  
Both have same amplitude and period but are shifted in time.  



### Why is a sinusoid periodic?  
It repeats identically after each period $T$.  



### Why is a sinusoid continuous?  
It is defined for all real time values without discontinuities.  



### Why is a sinusoid parametric, and with which parameters?  
It depends on amplitude ($A$), frequency ($f$), angular frequency ($\omega$), and phase ($\phi$).
